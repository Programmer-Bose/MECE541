**Log Tansformation of an Image without Scalar Multiplication**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

org_img = cv2.imread("man.jpg")
log_img = np.log(1+org_img)
prt_img = np.array(log_img,dtype = np.uint8)

plt.imshow(org_img)
plt.show()
plt.imshow(prt_img)
plt.show()
```

**Log Tansformation of an Image with Scalar Multiplication**

```python
image = cv2.imread('man.jpg')

c = 255 / np.log(1 + np.max(image))
log_image = c*(np.log(image + 1))

log_image = np.array(log_image, dtype = np.uint8)

# Display both images
plt.imshow(image)
plt.show()
plt.imshow(log_image)
plt.show()
```



 `image = cv2.imread('ron.jpeg')`
   - This line reads an image file named 'ron.jpeg' using the OpenCV library's `imread` function and stores it in a variable called 'image'. This line loads the image for further processing.

 `c = 255 / np.log(1 + np.max(image))`
   - This line calculates a constant 'c' used for contrast stretching. It finds the maximum pixel value in the 'image' and then computes 'c' as 255 divided by the natural logarithm (base e) of (1 + maximum pixel value in the image). This is done to enhance the image's contrast.

 `log_image = c*(np.log(image + 1))`
   - Here, a logarithmic transformation is applied to the 'image' data. Each pixel value is increased by 1 and then the natural logarithm is taken. Afterward, each result is multiplied by 'c'. 

 `log_image = np.array(log_image, dtype = np.uint8)`
   - This line converts the data type of 'log_image' to an unsigned 8-bit integer ('uint8'). This step is often done because images typically use 8-bit per channel representation with values ranging from 0 to 255. It ensures that the pixel values in 'log_image' are in the correct range for display and storage as an image.

   Then we are displaying both the images.

**Exponential Tansformation of an Image**

```python
k = 1
img_exp = (1+k)**org_img
plt.imshow(img_exp)
```


`k = 1`
   This line initializes a variable named `k` and assigns it the value 1. It seems to be setting a constant called "k."

`img_exp = (1+k)**org_img`
   In this line, a new variable `img_exp` is created. It calculates an "exponentiation" of each pixel value in the original image `org_img` to the power of `1+k`. This mathematical operation is applied to each pixel in the image.

`plt.imshow(img_exp)`
    To display the image `img_exp` that has undergone the previous mathematical transformation. It shows the transformed image, which have been modified based on the exponentiation operation with the value of `k`.

**Power Tansformation of an Image**

```python
c1 = 0.5
c2 = 1.2
img_pwr = (org_img**c2)
print(np.min(img_pwr),np.max(img_pwr))

plt.imshow(img_pwr)
cv2.imshow("Transformed Image", img_pwr)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

**Histogram of an Image**

```python
img_for_hist = cv2.imread('ron.jpeg',0)
histg_org = cv2.calcHist([img_for_hist], [0], None, [256], [0,256])

plt.plot(histg_org)
plt.show()
```

```python
equ_grey = cv2.equalizeHist(img_for_hist)
res_comp = np.hstack((img_for_hist,equ_grey))

hist_equ = cv2.calcHist([equ_grey], [0], None, [256], [0,256])

plt.plot(hist_equ)
plt.show()

cv2.imshow("result comparison",res_comp)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
