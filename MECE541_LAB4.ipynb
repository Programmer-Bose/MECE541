```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

org_img = cv2.imread('bird2.jpg')
org_img = cv2.cvtColor(org_img,cv2.COLOR_BGR2RGB)
```

### **Image Blurring using 3x3 Averaging Kernel**

```python
kernel5 = np.ones((5,5),np.float32)/25
res_avg5 = cv2.filter2D(org_img,-1,kernel5)

plt.subplot(121),plt.imshow(org_img),plt.title('Original')
plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(res_avg5),plt.title('Averaging(3x3)')
plt.xticks([]),plt.yticks([])
plt.show()
```

### **Image Blurring using 7x7 Averaging Kernel**

```python
kernel7 = np.ones((7,7),np.float32)/49
res_avg7 = cv2.filter2D(org_img,-1,kernel7)

plt.subplot(121),plt.imshow(org_img),plt.title('Original')
plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(res_avg7),plt.title('Averaging(7x7)')
plt.xticks([]),plt.yticks([])
plt.show()
```

### **Image Blurring using 9x9 Averaging Kernel**

```python
kernel9 = np.ones((9,9),np.float32)/81
res_avg9 = cv2.filter2D(org_img,-1,kernel9)

plt.subplot(121),plt.imshow(org_img),plt.title('Original')
plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(res_avg9),plt.title('Averaging(9x9)')
plt.xticks([]),plt.yticks([])
plt.show()
```

### **Image Blurring using Blur Function**

```python
res_blur = cv2.blur(org_img,(7,7))

plt.subplot(121),plt.imshow(org_img),plt.title('Original')
plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(res_blur),plt.title('Blurred')
plt.xticks([]),plt.yticks([])
plt.show()
```

### **Gaussian Blur**

```python
sample_a = cv2.imread('lena.jpg')
sample_a = cv2.cvtColor(sample_a,cv2.COLOR_BGR2RGB)

sample_b = cv2.imread('jet.png')
sample_b = cv2.cvtColor(sample_b,cv2.COLOR_BGR2RGB)

gaus_blur_a = cv2.GaussianBlur(sample_a,(9,9),0)
gaus_blur_b = cv2.GaussianBlur(sample_b,(9,9),0)

plt.subplot(121),plt.imshow(sample_a),plt.title('Sample A')
plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(gaus_blur_a),plt.title('Gaussian Blurred(9x9)')
plt.xticks([]),plt.yticks([])
plt.show()
plt.subplot(121),plt.imshow(sample_b),plt.title('Sample B')
plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(gaus_blur_b),plt.title('Gaussian Blurred(9x9)')
plt.xticks([]),plt.yticks([])
plt.show()
```

### **Median Blur**

```python
sample_c = cv2.imread('brain.jpg')
med_blur_c = cv2.medianBlur(sample_c,7)

plt.subplot(121),plt.imshow(sample_c),plt.title('Sample C')
plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(med_blur_c),plt.title('Median Blurred(7x7)')
plt.xticks([]),plt.yticks([])
plt.show()
```

### **Bilateral Filter**

```python
sample_d = cv2.imread('taj.jpg')
sample_d = cv2.cvtColor(sample_d,cv2.COLOR_BGR2RGB)

sample_e = cv2.imread('hill.jpg')
sample_e = cv2.cvtColor(sample_e,cv2.COLOR_BGR2RGB)

bilat_fil_d = cv2.bilateralFilter(sample_d,9,105,105)
bilat_fil_e = cv2.bilateralFilter(sample_e,9,105,105)

plt.subplot(121),plt.imshow(sample_d),plt.title('Sample D')
plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(bilat_fil_d),plt.title('Bilateral Filter')
plt.xticks([]),plt.yticks([])
plt.show()

plt.subplot(121),plt.imshow(sample_e),plt.title('Sample E')
plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(bilat_fil_e),plt.title('Bilateral Filter')
plt.xticks([]),plt.yticks([])
plt.show()
```
